```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


file_name ='pointing-up-2022-01-08 18_08_51.000.2opv8o75.json'
notebook_path = os.path.abspath("Notebook.ipynb")
pointing_up_file = os.path.join(os.path.dirname(notebook_path), file_name)
with open(pointing_up_file) as file:
    data = json.load(file)
    colum_names = []
    for sensor in data['payload']['sensors']:
        colum_names.append(sensor['name'])
    M = np.matrix(data['payload']['values'])
    df = pd.DataFrame(M, columns=colum_names)
    file.close()



display(df)
```

INDEX ACCELERATION



```python
df_i_acc_x = df[['indexAccX']]
print("Max value: "+ str(df_i_acc_x['indexAccX'].max()) + " metros/seg^2")
print("Min value: "+ str(df_i_acc_x['indexAccX'].min()) + " metros/seg^2")
```

```python
plt.figure();
df_i_acc_x.plot(figsize=(20,5));
```
